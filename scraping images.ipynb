```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

```python
url='https://www.passiton.com/inspirational-quotes'
data= urlopen(url)
data=data.read()
html_data=soup(data,'html.parser')
```

```python
images=html_data.find_all('div',{'id':'all_quotes'})[0]
images=images.find_all('div')
images.pop(0)
image=[]
for img in images:
    x=img.a.img
    image.append(x.attrs['src'])
c=0
for img in image:
    file='inspiration_quotes'+str(c)+'.jpg'
    c+=1
    with open(file,'wb') as f:
        image_data=urlopen(img)
        image_data=image_data.read()
        f.write(image_data)
```
